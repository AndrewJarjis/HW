```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/flying-etiquette-survey/flying-etiquette.csv')


age_counts = df['Age'].value_counts(normalize=True).sort_index()
age_counts *= 100 
plt.figure(figsize=(10,6))
plt.bar(x=age_counts.index, height=age_counts.values)
plt.title('Percentage Breakdown of Survey Participants by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Percentage of Participants')
plt.show()
```

```python

switch_counts = df['Is it rude to ask someone to switch seats with you in order to be closer to friends?'].value_counts()

plt.figure(figsize=(8,4))
plt.barh(switch_counts.index, switch_counts.values, color=['#FAC748', '#41A4E1', '#F44336'])
plt.title('Survey Respondents\' Opinions on Asking Someone to Switch Seats to be Closer to Friends')
plt.xlabel('Number of Respondents')
plt.show()

```
